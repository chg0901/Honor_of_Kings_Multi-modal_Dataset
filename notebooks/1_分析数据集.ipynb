```python
import pandas as pd
from pathlib import Path as p
import json
```

```python
# dataset_dir = p('/root/daji_RolePlay/Honor-of-Kings_RolePlay/datasets')
dataset_dir = p('../HoK_RolePlay/datasets')
dataset_dict = {}
for data_item in dataset_dir.glob('*.json'):
    try:
        cur_data_df = pd.read_json(data_item)
        dataset_dict[data_item.stem] = cur_data_df
        print(f'\n########## {data_item.stem} ##########')
        cur_data_df.info()
    except ValueError:
        cur_data_json = json.loads(data_item.read_text(encoding='utf-8'))
        dataset_dict[data_item.stem] = cur_data_json
        print(f'\n########## {data_item.stem} ##########')
        print(f'Warning: some slot are nan in {data_item.stem}')
        print('json_keys: ', cur_data_json.keys())
        print('json_len:', len(cur_data_json))
```
